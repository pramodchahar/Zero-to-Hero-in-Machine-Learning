# t-SNE

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
from sklearn.datasets import load_iris
from sklearn.manifold import TSNE
```

```python
iris = load_iris()
X_tsne = TSNE(learning_rate=100).fit_transform(iris.data)
```

t-SNE can help us to decide whether classes are separable in some linear or nonlinear representation. Here we can see that the 3 classes of the Iris dataset can be separated quite easily. They can even be separated linearly which we can conclude from the low-dimensional embedding of the PCA.

```python
plt.figure(figsize=(10, 10))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=iris.target,cmap=plt.cm.get_cmap('nipy_spectral',3))
plt.show()
```

## PCA and later t-sne 

```python
from sklearn.decomposition import PCA
```

```python
#load the dataset
from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape
```

```python
#PCA with only two components
digits_pca_2 = PCA(n_components=2)
```

```python
#fit PCA on the input data
digits_projected_2 = digits_pca_2.fit_transform(digits.data)
```

```python
digits_projected_2.shape
```

### Now fit t-sne

```python
X_tsne = TSNE(learning_rate=100).fit_transform(digits_projected_2)
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=digits.target,cmap=plt.cm.get_cmap('nipy_spectral',10))
plt.show()
```
